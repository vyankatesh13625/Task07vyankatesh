```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


data = load_breast_cancer()
X, y = data.data, data.target


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


linear_svm = SVC(kernel='linear')
rbf_svm = SVC(kernel='rbf')

linear_svm.fit(X_train, y_train)
rbf_svm.fit(X_train, y_train)

print("Linear Kernel Accuracy:", linear_svm.score(X_test, y_test))
print("RBF Kernel Accuracy:", rbf_svm.score(X_test, y_test))


scores = cross_val_score(rbf_svm, X_scaled, y, cv=5)
print("Cross-Validation Accuracy (RBF):", scores.mean())


params = {'C': [0.1, 1, 10], 'gamma': [1, 0.1, 0.01]}
grid = GridSearchCV(SVC(kernel='rbf'), param_grid=params, cv=5)
grid.fit(X_train, y_train)

print("Best Parameters (RBF):", grid.best_params_)
print("Best CV Accuracy:", grid.best_score_)


def plot_decision_boundary(model, X, y, title):
    h = .02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    model.fit(X, y)
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.xlabel('PCA 1')
    plt.ylabel('PCA 2')
    plt.title(title)
    plt.show()

plot_decision_boundary(SVC(kernel='linear'), X_pca, y, 'Linear SVM Decision Boundary')
plot_decision_boundary(SVC(kernel='rbf', gamma=0.1), X_pca, y, 'RBF SVM Decision Boundary')
```
